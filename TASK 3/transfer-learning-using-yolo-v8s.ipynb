Paths and class names


```python
 
train = '/kaggle/input/racecars/train/images'  
val = '/kaggle/input/racecars/valid/images'
test ='/kaggle/input/racecars/test/images'
names=['gt', 'nascar', 'open-wheel', 'prototype', 'rally']
```

Training the model


```python
results = model.train(
    data ='/kaggle/input/racecars/data.yaml',
    epochs = 50, 
    batch=16,
    imgsz=640,
    plots=True,
    name='yolov8supercar'
)
```

Importing new model made with trained weights

```python
new_model = YOLO('/kaggle/input/yolo-v8-supercars/pytorch/noice/1/best.pt')
```

Testing the new model


```python
results = model.predict(source=test, save=True, project='runs/detect/test_results', name='exp')
```

```python
metrics = new_model.val(data='/kaggle/input/racecars/data.yaml', split='test')  # Evaluate on the test set
print(metrics)
```
